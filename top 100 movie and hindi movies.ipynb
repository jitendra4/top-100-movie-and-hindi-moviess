```python
#importing the requried Libraries
from bs4 import BeautifulSoup
import requests
```

```python
page= requests.get("https://www.imdb.com/list/ls091520106/")
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
name_titles=[]
imdb_rating=[]
year_of_release=[]
```

```python
# let extract at the tags having the name-titles
titles=soup.find_all('div',class_='lister-item mode-detail')
titles[0:4]
```

```python
for i in titles:
    name_titles.append(i.get_text().replace("\n",""))
name_titles
```

```python
titles[0].get_text().replace("\n","")
```

```python
# let extract at the tags having the year-of-release
release=soup.find_all('span',class_="lister-item-year text-muted unbold")
release[0:4]
```

```python
for i in release:
    year_of_release.append(i.get_text().replace("\n",""))
year_of_release
```

```python
release[0].get_text().replace("\n","")
```

```python
# let extract at the tags having the imdb rating
rating=soup.find_all('span',class_="ipl-rating-star__rating")
rating[0:4]
```

```python
for i in rating:
    imdb_rating.append(i.get_text().replace("\n",""))
imdb_rating
```

```python
rating[0].get_text().replace("\n","")
```

```python
print(len(name_titles),
      len(imdb_rating),
      len(year_of_release))
```

```python
import pandas as pd
movies=pd.DataFrame({})
movies['titles']=name_titles
movies['rating']=imdb_rating
movies
```
